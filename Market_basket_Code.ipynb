```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
```

```python
data = pd.read_csv("teleco_market_basket.csv")
```

```python
data.head()
```

```python
data = data[~data.Item01.isna()]
```

```python
data.head()
```

```python
def make_list(data):
    l1 = []
    for i in data:
        if i != "0":
            l1.append(i)
    return l1
```

```python
#test = [make_list(i[1].fillna("0")) for i in data.iterrows()]
```

```python
data['all_items'] = [make_list(i[1].fillna("0")) for i in data.iterrows()]
```

```python
data.head()
```

```python
encoder = TransactionEncoder().fit(data['all_items'])
onehot = encoder.transform(data['all_items'])
onehot = pd.DataFrame(onehot, columns= encoder.columns_)
```

```python
onehot.head()
```

```python
onehot.to_csv('onehot.csv')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(onehot, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)
rules
```

```python
key_col = ['antecedents','consequents', 'support', 'confidence', 'lift', 'leverage']
#rule_three = rules[key_col].sort_values('lift', ascending = False)[:3]
```

```python
rules[key_col].sort_values('lift', ascending = False).head(4)
```

```python
from pandas.plotting import parallel_coordinates
rules['antecedents'] = rules['antecedents'].apply(lambda antecedent: list(antecedent)[0])
rules['consequents'] = rules['consequents'].apply(lambda consequents: list(consequents)[0])
rules['rule'] = rules.index
coords = rules[['antecedents', 'consequents', 'rule']]
```

```python
parallel_coordinates(coords, 'rule')
```
